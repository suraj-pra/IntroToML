```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../input/up-crime-data-year-2019/MR data - Compiled Data Set.csv")
df.head()
```

```python
df = df.drop(['Title'], axis= 1)
df = df.drop(['Text'], axis = 1)
df = df.fillna(0)
df.head()
```

```python
df.columns = [i for i in range(len(df.columns))]
df.head()
```

```python
reason = [i for i in df[1] if i != 0]
reason = list(dict.fromkeys(reason))
reason
```

```python
dict_ = {0:0}
for i in range(len(reason)):
    dict_[reason[i]] = i + 1

df[1] = df[1].apply(lambda x: dict_[x])
df.head()
```

```python
s = df[0].value_counts().sort_index()

plt.bar(s.index, s.values, color = '#ffa31a', width = 0.3)

plt.xlabel("City")
plt.ylabel("Total Crimes")
plt.title("City Wise Crimes")
plt.show()
```

```python
df[11] = 0
df[11] = df[2] + df[3]
df[11] = df[11] + df[4]
df.head()
```

```python
s = df.groupby(by = 1).sum()[11]
s = s.drop(0);

plt.figure(figsize=(17.5,7.5))
plt.rcParams.update({'font.size': 13})
plt.bar(reason, s.values, color = '#66ff33', width = 0.3)

plt.xlabel("Reasons")
plt.ylabel("Total Murder Victims")
plt.title("Murders")
plt.show()
```

```python
s = df[11].value_counts().sort_index()

plt.bar((s.index.astype(int)).astype(str), s.values, color = '#ff33cc', width = 0.6)

plt.xlabel("No. Of Murders")
plt.ylabel("Total Crimes")
plt.title("No. of People Wise")
plt.show()
```

```python
df[[12,13,14]] = df[8].str.split(n = 2, pat = ",", expand = True).apply(pd.to_numeric)
df = df.fillna(0)
df.head()
```

```python
s0 = df[12].value_counts().sort_index()
s1 = df[13].value_counts().sort_index()
s2 = df[14].value_counts().sort_index()

s = s0.add(s1, fill_value = 0)
s = s.add(s2, fill_value = 0)

s = s.drop(0)

plt.figure(figsize=(10,5))
plt.rcParams.update({'font.size': 10})
plt.bar((s.index.astype(int)).astype(str), s.values, color = '#ff4d4d', width = 0.5)

plt.xlabel("SECTION")
plt.ylabel("Total Crimes Against Women")
plt.title("Crime Against Women")
plt.show()
```
